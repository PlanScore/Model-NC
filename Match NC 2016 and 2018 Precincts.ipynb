```python
import geopandas, os
os.environ['DYLD_LIBRARY_PATH'] = './.venv-NC/lib'
```

```python
geogs_2016 = geopandas.read_file('NC-Geographies.gpkg', layer='precincts')
geogs_2016
```

```python
geogs_2018 = geopandas.read_file('/Users/migurski/Downloads/NCSBE-2016-2018/NC_G18.shp').to_crs(geogs_2016.crs)
geogs_2018
```

```python
help(geopandas.sjoin)

geogs_joined = geopandas.sjoin(geogs_2018, geogs_2016, op='intersects', how='left')

geogs_joined.shape
```

```python
core_cols = ['COUNTY_NAM', 'PREC_ID', 'geometry', 'index_right', 'county_name', 'precinct_id', 'psid']

geogs_matched = geogs_joined[geogs_joined.county_name == geogs_joined.COUNTY_NAM]

PSIDs_2018 = {k: None for k in geogs_2018.index}

for (index_2018, row_joined) in geogs_matched.iterrows():
    geom_2018 = row_joined.geometry
    geom_2016 = geogs_2016.loc[row_joined.index_right].geometry
    try:
        union = geom_2018.union(geom_2016)
        intersection = geom_2018.intersection(geom_2016)
    except:
        continue
    if intersection.area / union.area < .95:
        continue
    PSIDs_2018[index_2018] = str(row_joined.psid)
    #print(index_2018, row_joined.index_right, row_joined.psid)

geogs_2018['PSID'] = geogs_2018.index.map(PSIDs_2018)
geogs_2018
```

```python
geogs_2018.to_file('geogs_2018.shp')
```
